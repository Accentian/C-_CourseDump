```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# Read the data
piezo = pd.read_csv('piezo.csv')
piezo
```

```python
#Call predictor variables and rows as capital X. Might not be feasible here.
X = piezo.iloc[:,0:5]

#Get all the rows and columns total, save to y. Testing specific column here.
y = piezo.loc[:,"volume"]

#The import function comes into play here
#For the Test set:
#This line pass in the dataframe with all the predictor variables (X), then the outcome variables (y), then the percentage size
train_test_split(X,y,test_size=.2)
```

```python
#Save the outputs. As shown above, we recieve the training datasets first followed by the testing datasets
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2)

#Seed will give you the same split every time
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2, random_state = 1337)
```
