```python
import numpy as np
from sklearn import datasets
iris_X, iris_y = datasets.load_iris(return_X_y=True)
np.unique(iris_y)

from sklearn import linear_model
```

```python
X = np.c_[ .5, 1].T
y = [.5, 1]
test = np.c_[ 0, 2].T
regr = linear_model.LinearRegression()
```

```python
import matplotlib.pyplot as plt 
plt.figure() 

np.random.seed(0)
for _ in range(10): 
    this_X = .4 * np.random.normal(size=(2, 1)) + X
    regr.fit(this_X, y)
    plt.plot(test, regr.predict(test)) 
    plt.scatter(this_X, y, s=1000)  
```

```python
# A solution in high-dimensional statistical learning is to shrink the regression coefficients to zero: any two randomly 
# chosen set of observations are likely to be uncorrelated. This is called Ridge regression:
regr = linear_model.Ridge(alpha=.8)
plt.figure()
np.random.seed(0)
for _ in range(4): 
    this_X = .9 * np.random.normal(size=(2, 1)) + X
    regr.fit(this_X, y)
    plt.plot(test, regr.predict(test)) 
    plt.scatter(this_X, y, s=50) 
```
